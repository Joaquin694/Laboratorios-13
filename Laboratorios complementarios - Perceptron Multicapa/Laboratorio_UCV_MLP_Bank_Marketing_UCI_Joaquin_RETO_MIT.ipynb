Importar Librerias

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

Cargar dataset

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

Escalar Datos

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Dividir los datos en entrenamiento (70%) y prueba (30%)

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

Definir al menos TRES arquitecturas diferentes de Perceptrón Multicapa

```python
arquitecturas = {
    "Modelo 1 (Simple)": (5,),
    "Modelo 2 (Intermedio)": (20, 10),
    "Modelo 3 (Complejo)": (100, 100, 50)
}

resultados = []
```

Entrenar y comparar cada arquitectura

```python
for nombre, capas in arquitecturas.items():
    mlp = MLPClassifier(hidden_layer_sizes=capas, max_iter=2000, random_state=42)
    mlp.fit(X_train, y_train)

    y_pred = mlp.predict(X_test)

    resultados.append({
        "Arquitectura": nombre,
        "Capas Ocultas": str(capas),
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1-Score": f1_score(y_test, y_pred)
    })
```

Mostrar los resultados en una tabla comparativa

```python
df_resultados = pd.DataFrame(resultados)
print("--- RESULTADOS DEL EXPERIMENTO ---")
display(df_resultados)
```
